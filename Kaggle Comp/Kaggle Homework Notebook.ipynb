## Feature importance: quantitative variables

Idea courtesty of @rawashishsin on Kaggle

```python
#first use correlation matrix to check whether there are any redundant/highly correlated numeric variables (idea courtesty of @wesleyhuan on Kaggle)

numeric_cols = training.select_dtypes(include=["number"]).columns.tolist()
numeric_cols = [col for col in numeric_cols if col not in ["id"]]

corr_matrix = training[numeric_cols].corr()

fig, ax = plt.subplots(figsize=(20, 10))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", ax=ax)
ax.set_title("Correlation Matrix of Numeric Features")
plt.show()
```

There is no high correlation between any of the numeric variables, so I will not be dropping any of these variables as there is no clear redundancy.

```python
#run ANOVA F-test for all numeric features
X_num = training[numeric_cols].copy()
y = training["Irrigation_Need"].copy()

f_stats, p_vals = f_classif(X_num, y)

anova_df = pd.DataFrame({
    "feature": numeric_cols,
    "f_stat": f_stats,
    "p_value": p_vals
}).sort_values("f_stat", ascending=False).reset_index(drop=True)

print(anova_df.head(10))
```

```python
top_features = [
    "Soil_Moisture",
    "Wind_Speed_kmh",
    "Temperature_C",
    "Rainfall_mm"
]

plot_data = training[["Irrigation_Need"] + top_features].copy()

plot_data["Irrigation_Need"] = pd.Categorical(
    plot_data["Irrigation_Need"],
    categories=["Low", "Medium", "High"],
    ordered=True
)

plot_data = plot_data.melt(
    id_vars="Irrigation_Need",
    value_vars=top_features,
    var_name="feature",
    value_name="value"
)

(
    ggplot(plot_data, aes(x="Irrigation_Need", y="value"))
    + geom_boxplot()
    + facet_wrap("~feature", scales="free_y", ncol=2)
    + labs(
        title="Top Quantitative Features by Irrigation Need",
        x="Irrigation Need",
        y="Value"
    )
    + theme_minimal()
    + theme(
        figure_size=(10, 8),
        axis_text_x=element_text(rotation=0)
    )
)
```

## Feature importance: categorical variables

Courtesty of @rawashishsin

```python
#select categorical variables only
categorical_cols = training.select_dtypes(exclude=["number"]).columns.tolist()
categorical_cols = [col for col in categorical_cols if col != "Irrigation_Need"]
categorical_cols
```

```python
#run chi^2 for all categorical features
X_cat = training[categorical_cols].copy()
y = training["Irrigation_Need"].copy()

chi2_results = []

for col in categorical_cols:
    contingency_table = pd.crosstab(X_cat[col], y)
    chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

    chi2_results.append({
        "feature": col,
        "chi2_stat": chi2_stat,
        "p_value": p_val
    })

chi2_df = pd.DataFrame(chi2_results) \
    .sort_values("chi2_stat", ascending=False) \
    .reset_index(drop=True)

print(chi2_df.head(10))
```

```python
top_cat_features = [
    "Crop_Growth_Stage",
    "Mulching_Used",
    "Water_Source",
    "Irrigation_Type",
    "Crop_Type"
]

plot_data = training[top_cat_features + ["Irrigation_Need"]].copy()

plot_data["Irrigation_Need"] = pd.Categorical(
    plot_data["Irrigation_Need"],
    categories=["Low", "Medium", "High"],
    ordered=True
)

plot_data = plot_data.melt(
    id_vars="Irrigation_Need",
    value_vars=top_cat_features,
    var_name="feature",
    value_name="category"
)

(
    ggplot(plot_data, aes(x="category", fill="Irrigation_Need"))
    + geom_bar(position="fill")
    + facet_wrap("~feature", scales="free_x")
    + labs(
        title="Top Categorical Features by Irrigation Need",
        x="Category",
        y="Proportion"
    )
    + theme_minimal()
    + theme(
        figure_size=(12, 8),
        axis_text_x=element_text(rotation=45, ha="right")
    )
)
```

The most important numerical variable appears to be soil moisture due to having the largest F-statistic. The most important categorical variable appears to be crop growth stage due to having the largest chi-square statistic. The visuals show clear separation between low, medium, and high irrigation need for these variables.

## Baseline modeling

```python
# define X and y
X = training.drop(["Irrigation_Need", "id"], axis=1)
cat_columns = ['Soil_Type',
 'Crop_Type',
 'Crop_Growth_Stage',
 'Season',
 'Irrigation_Type',
 'Water_Source',
 'Mulching_Used',
 'Region']
X[cat_columns] = X[cat_columns].astype("category")
y = training["Irrigation_Need"]
```

```python
X.head()
```

```python
y.head()
```

```python
#define column transformer that dummifies categorical variables
ct_dummify = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown='ignore'),
    make_column_selector(dtype_include=["object", "category"]))],
  remainder = "passthrough"
).set_output(transform="pandas")

ct_dummify.fit_transform(X).head()
```

```python
#define baseline models (with multi-class imbalance addressed)

## Bagging models

#baseline random forest pipeline
rf_pipeline = Pipeline(
    [("preprocessing", ct_dummify),
    ("rf", RandomForestClassifier(random_state=321, class_weight='balanced', n_jobs = -1))]
)

#baseline extra trees classifier pipeline
et_pipeline = Pipeline(
    [("preprocessing", ct_dummify),
    ("et", ExtraTreesClassifier(random_state=321, class_weight='balanced', n_jobs = -1))]
)

## Boosting models

#baseline xgboost model
xgb_model = XGBClassifier(random_state = 321, enable_categorical=True)

#baseline histgradient boosting model
hgb_model = HistGradientBoostingClassifier(random_state=321, class_weight='balanced', categorical_features="from_dtype")
```

```python
#define cross-validation strategy
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=321)
```

```python
#random forest default model with 5-fold cross validation
cv_scores_rf = cross_val_score(rf_pipeline, X, y, cv=cv, scoring="balanced_accuracy", n_jobs=-1)
print(cv_scores_rf)
print(cv_scores_rf.mean())
print(cv_scores_rf.std())
cross_val_predict_rf = cross_val_predict(rf_pipeline, X, y, cv=cv, n_jobs=-1)
ConfusionMatrixDisplay.from_predictions(y, cross_val_predict_rf)
```

```python
#extra trees default model with 5-fold cross validation
cv_scores_et = cross_val_score(et_pipeline, X, y, cv=cv, scoring="balanced_accuracy", n_jobs=-1)
print(cv_scores_et)
print(cv_scores_et.mean())
print(cv_scores_et.std())
cross_val_predict_et = cross_val_predict(et_pipeline, X, y, cv=cv, n_jobs=-1)
ConfusionMatrixDisplay.from_predictions(y, cross_val_predict_et)
```

```python
#xgboost default model with 5-fold cross validation

#encode outcome variable as numeric for xgboost
y_encoded = y.map({"Low": 0, "Medium": 1, "High": 2})
cv_scores_xgb = cross_val_score(xgb_pipeline, X, y_encoded, cv=cv, scoring="balanced_accuracy", n_jobs=-1)
print(cv_scores_xgb)
print(cv_scores_xgb.mean())
print(cv_scores_xgb.std())
cross_val_predict_xgb = cross_val_predict(xgb_pipeline, X, y_encoded, cv=cv, n_jobs=-1)
ConfusionMatrixDisplay.from_predictions(y_encoded, cross_val_predict_xgb)
```

```python
#histgradient boosting default model with 5-fold cross validation
cv_scores_hgb = cross_val_score(hgb_pipeline, X, y, cv=cv, scoring="balanced_accuracy", n_jobs=-1)
print(cv_scores_hgb)
print(cv_scores_hgb.mean())
print(cv_scores_hgb.std())
cross_val_predict_hgb = cross_val_predict(hgb_pipeline, X, y, cv=cv, n_jobs=-1)
ConfusionMatrixDisplay.from_predictions(y, cross_val_predict_hgb)
```

## Model tuning

```python
#random forest model tuning
param_grid_rf = {
    "rf__n_estimators": [100, 200, 300],
    "rf__max_features": ["sqrt", "log2"],
    "rf__max_depth": [None, 10, 20],
    "rf__min_samples_leaf": [1, 2, 4]
}

grid_search_rf = GridSearchCV(
    rf_pipeline,
    param_grid_rf,
    cv=cv,
    scoring="balanced_accuracy",
    n_jobs=-1
)

grid_search_rf.fit(X_best, y)

print("Best balanced accuracy score:", grid_search_rf.best_score_)
print("Best parameters:", grid_search_rf.best_params_)
```
