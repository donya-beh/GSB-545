```python
import pandas as pd
import numpy as np
import sweetviz as sv
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# read in data
credit_card = pd.read_csv("C:\\Users\\donya\\Downloads\\creditcard.csv")
credit_card.head()
```

#1

```python
# EDA with sweetviz
report = sv.analyze(credit_card)
report.show_html("sweetviz_report.html")
```

Highlights from EDA
- Limit balance skewed to the right with a median of $140,000
- Age is skewed to the right with a median of 34 years old. All individuals in this data set are > 18 years old
- All bill amounts are skewed to the right
- The outcome variable, default payment next month, is imbalanced (individuals primarily will not default on their payment next month)

```python
# define X and y
X = credit_card.drop("default payment next month", axis=1)
y = credit_card["default payment next month"]

# split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=321)

#training random forest model (baseline)
rf = RandomForestClassifier(random_state=321)
rf.fit(X_train, y_train)

#evaluating random forest model
y_pred_rf = rf.predict(X_test)
print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))

#with cross validation
cv_scores_rf = cross_val_score(rf, X, y, cv=5)
print("Random Forest Cross-Validation Scores:", cv_scores_rf)
print("Random Forest Average CV Score:", np.mean(cv_scores_rf))
print("Random Forest CV Score Standard Deviation:", np.std(cv_scores_rf))

#training xgboost model (baseline)
xgb = XGBClassifier(random_state=321, use_label_encoder=False, eval_metric='logloss')
xgb.fit(X_train, y_train)

#evaluating xgboost model
y_pred_xgb = xgb.predict(X_test)
print("XGBoost Classification Report:")
print(classification_report(y_test, y_pred_xgb))
print("XGBoost Accuracy:", accuracy_score(y_test, y_pred_xgb))

#with cross validation
cv_scores_xgb = cross_val_score(xgb, X, y, cv=5)
print("XGBoost Cross-Validation Scores:", cv_scores_xgb)
print("XGBoost Average CV Score:", np.mean(cv_scores_xgb))
print("XGBoost CV Score Standard Deviation:", np.std(cv_scores_xgb))
```

#2
- Both models perform well in terms of accuracy (~ 0.80 for each). However it seems like due to the imbalance in the dataset, the models' high accuracy scores are likely being dominated by correct classifications of individuals who will not default on their payment the next month (outcome = 0). We can see this in the lower recall and precision scores for both models for the outcome = 0 group compared to the outcome = 1 (defaults on payment the next month) group

- The average accuracy scores and standard deviation for the cross-validated models are very similar between both the Random Forest and XGBoost models. We see accuracy scores that are consistently around 0.80 for both models with not much deviation. Due to the imbalanced data, we are likely obtaining folds of data that are primarily representing individuals who do not default on the next month's payment. This may explain the small deviation in accuracy score across the five folds. The model is consistently performing well at classifying our default = 1 cases.

#4

```python
#adjust for class imbalance in random forest model
rf_balanced = RandomForestClassifier(random_state=321, class_weight='balanced')
rf_balanced.fit(X_train, y_train)
y_pred_rf_balanced = rf_balanced.predict(X_test)
print("Random Forest (Balanced) Classification Report:")
print(classification_report(y_test, y_pred_rf_balanced))

#adjust for class imbalance in xgboost model
xgb_balanced = XGBClassifier(random_state=321, use_label_encoder=False, eval_metric='logloss', scale_pos_weight=10)
xgb_balanced.fit(X_train, y_train)
y_pred_xgb_balanced = xgb_balanced.predict(X_test)
print("XGBoost (Balanced) Classification Report:")
print(classification_report(y_test, y_pred_xgb_balanced))
```

- The accuracy did not change for the random forest balanced model, but the precision score increased for classification of individuals who will default on the next month's payment. However, the recall score decreased meaning that the model struggled with correctly classifying individuals who default correctly.

- Interestingly, we see accuracy decrease when accounting for class imbalance with XGBoost. The classification report of this model is quite different from the random forest balanced model. In this model, the recall increased substantially but decreased quite a bit as well in terms of precision. This indicates that of individuals who actually defaulted, the model was able to capture 74% (high accuracy) of them but there are many false positives-individuals who were classified as defaulting but they actually did not default (low precision).

```python
#XGBoost model with subsample feature
xgb_subsample = XGBClassifier(random_state=321, use_label_encoder=False, eval_metric='logloss', subsample=0.8)
xgb_subsample.fit(X_train, y_train)
y_pred_xgb_subsample = xgb_subsample.predict(X_test)
print("XGBoost (Subsample) Classification Report:")
print(classification_report(y_test, y_pred_xgb_subsample))
```

- The model's accuracy is close to the original XGBoost model without any subsampling

- Both precision and recall have increased for the defaulted = 1 outcome compared to the original XGBoost model. This indicates that the model with subsampling is better able to classify individuals who default on their payments in the next month

#6

```python
#tuning 2 hyperparameters with 2 values for Random Forest using imbalance adjustments
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10]
}
grid_rf = GridSearchCV(estimator=rf_balanced, param_grid=param_grid_rf, cv=5, n_jobs=-1)
grid_rf.fit(X_train, y_train)
print("Best Hyperparameters for Random Forest:", grid_rf.best_params_)
best_rf = grid_rf.best_estimator_
y_pred_best_rf = best_rf.predict(X_test)
print("Tuned Random Forest Classification Report:")
print(classification_report(y_test, y_pred_best_rf))

#tuning 2 hyperparameters with 2 values for XGBoost using imbalance adjustments
param_grid_xgb = {
    'n_estimators': [100, 200],
    'max_depth': [3, 6]
}
grid_xgb = GridSearchCV(estimator=xgb_balanced, param_grid=param_grid_xgb, cv=5, n_jobs=-1)
grid_xgb.fit(X_train, y_train)
print("Best Hyperparameters for XGBoost:", grid_xgb.best_params_)
best_xgb = grid_xgb.best_estimator_
y_pred_best_xgb = best_xgb.predict(X_test)
print("Tuned XGBoost Classification Report:")
print(classification_report(y_test, y_pred_best_xgb))
```

- The random forest model's accuracy score is the same as the default random forest's accuracy score. However, we see that this model's precision score for default = 1 has increased compared to the default model but also decreased in terms of its recall. Therefore, of everyone the model predicted as defaulting on next month's payment, the model is able to capture a 71% of individuals who actually did default, but still incorrectly classifying those who defaulted as individuals who did not default.

- The accuracy score for the balanced XGBoost model has decreased compared to the default model. The precision score for default = 1 has decreased, while the recall score for this group has increased. This indicates that 64% of individuals are classified correctly as defaulting, but of all the individuals the model predicted would default, only 41% actually defaulted.
